## Knapsack Problem

```python
def knapSack(W, wt, val, n):
    K = [[0 for x in range(W + 1)] for x in range(n + 1)]
 
    for i in range(n + 1):
        for w in range(W + 1):
            if i == 0 or w == 0:
                K[i][w] = 0
            elif wt[i-1] <= w:
                K[i][w] = max(val[i-1]
                          + K[i-1][w-wt[i-1]], 
                              K[i-1][w])
            else:
                K[i][w] = K[i-1][w]
 
    return K[n][W]
```

```python
wt = [12, 2, 1, 1, 4]
val = [4, 2, 2, 1, 10]
W = 15
n = len(val)
print(knapSack(W, wt, val, n))
```

```python
import numpy as np
import itertools
def Knapsack_Counting(S,W,x):
  soluciones=0
  multi=np.multiply(S,x)
  sumk=np.sum(multi)
  if(sumk <= W):
    soluciones+=1
  return soluciones
```

```python
S=np.array([1,1,2,4,12])
W=15
x = list(itertools.product([0,1], repeat=len(S)))
solu=0
for i in list(x): 
  solu+=Knapsack_Counting(S,W,np.array(i))

print(solu)
```

Count and calculate the exact proportion of  “Knapsack solutions.” for the problem in the image

```python
S = np.array([1,1,2,4,12]) 
W = 15 
soluciones = 0 
p = 0 

x = list(itertools.product([0,1], repeat=S.size))
 
for i in list(x):   
  soluciones+=Knapsack_Counting(S,W,np.array(i))

p = soluciones/(2**S.size)
print('Soluciones: ',(soluciones))
print('Proporción: ',(p*100),"%")
print('Combinaciones totales: ',(2**S.size))
```

###Estimate the proportion and number of “Knapsack solutions.” if the sizes are (1,2,3,4,5,6,7,9,10) and the capacity of the knapsack is 10 using Mote Carlo with 1.000, 10.000 random binary vectors

```python
S = np.array([1,2,3,4,5,6,7,9,10]) 
W = 10
soluciones = 0 
p = 0 

 
runs=1000
for i in range(0,runs):  
  x = np.random.randint(2, size=S.size) 
  soluciones+=Knapsack_Counting(S,W,x)

p = soluciones/runs
print('Soluciones: ',(soluciones))
print('Proporción: ',(p*100),"%")
```
